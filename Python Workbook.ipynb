**pandas DataFrame**

DataFrames can be created from **lists**, **dictionaries**, and pandas **Series**

```python
# step_data = [3620, 7891, 9761, 3907, 4338, 5373]
```

```python
## Cycling distance
cycling_data = [10.7, 0, None, 2.4, 15.3, 10.9, 0, None]
```

Create a list of tuples  
**Tuple**: Round brackets ( )

```python
joined_data = list(zip(step_data, cycling_data))
print(joined_data)
```

We've just created a list of tuples!

**pandas DataFrame**

```python
activity_df = pd.DataFrame(joined_data)
print(activity_df)
```

We can add column names to DataFrame and an index

```python
activity_df = pd.DataFrame(joined_data, index=pd.date_range('20150329', periods=6), columns=['Walking','Cycling'])
print(activity_df)
```

DataFrame rows can be indexed by row using the 'loc' and ' iloc ' methods

Select row of data by index name

```python
print(activity_df.loc['2015-04-01'])
```

Select row of data by integer position

```python
print(activity_df.iloc[-3])
```

**DataFrame columns can be indexed by name**

1. Dictionary approach

```python
print(activity_df['Walking'])
```

2. Object-oriented approach

```python
print(activity_df.Cycling)
```

DataFrame columns can also be indexed by integer

```python
# First column (Column 0)
print(activity_df.iloc[:, 0])
```

## Plotting graphs using Matplotlib & Seaborn

**Matplotlib** allows Python to plot graphs as pretty as MATLAB does

Exercise 1: Lines

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 50)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, 'r*-', label='sin')
plt.plot(x, y_cos, 'bo-', label='cos')
plt.xlabel('x coordinates')
plt.ylabel('y coordinates')
plt.title('A Matplotlib example')
plt.legend(loc='best')
plt.grid(axis='both')
plt.show()
```

**Seaborn** is an alternative to Matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 50)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create a DataFrame for Seaborn
import pandas as pd
df = pd.DataFrame({'x': x, 'sin': y_sin, 'cos': y_cos})

# Plot using Seaborn
sns.lineplot(x='x', y='sin', data=df, label='sin', marker='*', linestyle='-')
sns.lineplot(x='x', y='cos', data=df, label='cos', marker='o', linestyle='-')

plt.xlabel('x coordinates')
plt.ylabel('y coordinates')
plt.title('A Seaborn example')
plt.legend(loc='best')
plt.grid(axis='both')
plt.show()
```

Exercise 2: Scatter plot using **matplotlib**

```python
pts = np.random.randint(0, 20, (10, 2))

plt.scatter(pts[:, 0], pts[:, 1], color='b', marker='*')
plt.xlabel('x coord')
plt.ylabel('y coord')
plt.title('A Matplotlib scatter plot example')
plt.grid(axis='both')
plt.show()
```

Exercise 3: Histogram using **Matplotlib**

```python
values = np.random.randint(0, 20, 100)
bins_1 = np.arange(0, 21)

plt.hist(values, color='b', bins=bins_1)
plt.xlabel('Values')
plt.ylabel('Frequency of occurence')
plt.title('A Matplotlib histogram example')
plt.xticks(range(0, 20))
plt.grid(axis='y')
plt.show()
```

Exercise 4: Bar Chart using **Matplotlib**

```python
x = ['A', 'B', 'C', 'F']
y = [21, 42, 29, 8]

plt.bar(x, y, color=['b', 'g', 'g', 'r'])
plt.xlabel('Grade')
plt.ylabel('No. of students')
plt.title('Summary of results')
plt.show()
```

Exercise 5: Subplot using **Matplotlib**

```python
x = np.linspace(-np.pi, np.pi, 25)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y_sin, 'r.-')
plt.title('Sine')

plt.subplot(2, 1, 2)
plt.plot(x, y_cos, 'b^-')
plt.title('Cosine')

plt.tight_layout()
plt.show()
```
